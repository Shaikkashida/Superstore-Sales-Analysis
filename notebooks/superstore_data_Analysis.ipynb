```python
df=df.drop_duplicates()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df['Order Month']=df['Order Date'].dt.month
```

```python
df['Order Year']=df['Order Date'].dt.year
```

```python
df['Order Month-Year']=df['Order Date'].dt.to_period('M')
```

```python
df['Profit Margin']=(df['Profit']/df['Sales'])*100
```

```python
df[['Order Date', 'Order Month', 'Order Year', 'Order Month-Year','Profit Margin']].head(5)
```

```python
df.dtypes
```

```python
total_sales=df['Sales'].sum()
total_sales
```

```python
total_profit=df['Profit'].sum()
total_profit
```

```python
total_quantity=df['Quantity'].sum()
total_quantity
```

```python
yearly_trend=df.groupby('Order Year')[['Sales','Profit']].sum()
yearly_trend
```

```python
yearly_trend.plot(kind='line',figsize=(8,5),title='Yearly sales & profit trend')
```

```python
category_performance=df.groupby('Category')[['Sales','Profit']].sum().sort_values(by='Sales' , ascending=False)
category_performance
```

```python
category_performance.plot(kind='bar',figsize=(8,5),title='Category-wise sales & profit')
```

```python
sub_category_perf=df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales' , ascending=False)
sub_category_perf
```

```python
Region_Perf=df.groupby('Region')[['Sales','Profit']].sum()
Region_Perf
```

```python
Region_Perf.plot(kind='bar',figsize=(7,4), title='Region-wise sales & profit')
```

```python
segment_perf=df.groupby('Segment')[['Sales','Profit']].sum()
segment_perf
```

```python
df.groupby('Category')['Profit Margin'].mean().sort_values(ascending=False)
```
